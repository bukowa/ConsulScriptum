```python
import numpy as np
from sklearn.linear_model import LinearRegression

# ==============================================================================
# 1. TRAINING FUNCTION (Now with duplicate removal)
# ==============================================================================
def train_model(training_data_string):
    """
    Parses unique training data points, performs linear regression, and
    returns the calculated model parameters.
    """
    logical_x, logical_y, display_x, display_y = [], [], [], []

    # Get unique lines to prevent duplicate data points from skewing the model
    unique_lines = set(training_data_string.strip().split('\n'))

    for line in unique_lines:
        parts = line.split(',')
        if len(parts) == 4:
            try:
                logical_x.append(float(parts[0]))
                logical_y.append(float(parts[1]))
                display_x.append(float(parts[2]))
                display_y.append(float(parts[3]))
            except ValueError:
                continue

    if not display_x:
        print("Error: No valid training data found.")
        return None

    lx, ly = np.array(logical_x), np.array(logical_y)
    dx, dy = np.array(display_x).reshape(-1, 1), np.array(display_y).reshape(-1, 1)

    model_x = LinearRegression().fit(dx, lx)
    model_y = LinearRegression().fit(dy, ly)

    return {
        'scale_x': model_x.coef_[0],
        'offset_x': model_x.intercept_,
        'scale_y': model_y.coef_[0],
        'offset_y': model_y.intercept_
    }

# ==============================================================================
# 2. VERIFICATION FUNCTION (Now with duplicate removal)
# ==============================================================================
def verify_model(test_data_string, model_params):
    """
    Verifies the accuracy of a given model against a unique set of test data.
    """
    total_points = 0
    correct_matches = 0

    scale_x, offset_x = model_params['scale_x'], model_params['offset_x']
    scale_y, offset_y = model_params['scale_y'], model_params['offset_y']

    # Get unique lines for a more meaningful accuracy report
    unique_lines = set(test_data_string.strip().split('\n'))

    for line in unique_lines:
        parts = line.split(',')
        if len(parts) != 4: continue

        try:
            true_lx, true_ly = int(parts[0]), int(parts[1])
            display_x, display_y = float(parts[2]), float(parts[3])
        except ValueError:
            continue

        total_points += 1

        pred_lx_float = (scale_x * display_x) + offset_x
        pred_ly_float = (scale_y * display_y) + offset_y
        pred_lx = int(round(pred_lx_float))
        pred_ly = int(round(pred_ly_float))

        if (pred_lx, pred_ly) == (true_lx, true_ly):
            correct_matches += 1

    print("--- Verification Report (on unique data points) ---")
    if total_points == 0:
        print("No valid data points found in the test data.")
        return

    accuracy = (correct_matches / total_points) * 100
    print(f"Unique Points Checked: {total_points}")
    print(f"Correct Matches:       {correct_matches}")
    print(f"Accuracy:              {accuracy:.2f}%")
    print("-----------------------------------------------------")

# --- SCRIPT EXECUTES FROM HERE ---
print("--- Starting the automated workflow ---")

model_params = train_model(training_data)

if model_params:
    print("\nModel training complete (using unique data points). Parameters found:")
    for key, value in model_params.items():
        print(f"  {key}: {value:.4f}")

    print("\nVerifying the model on the test data...")
    verify_model(test_data, model_params)
else:
    print("\nTraining failed. Cannot proceed to verification.")
```
